```python
import main
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def square_function(x_start, x_stop):
    x = np.linspace(x_start,x_stop,100)
    result = []
    for x_ in x:
        result.append(main.fun1(x_))
    plt.figure()
    plt.plot(x, result)
    plt.title(f"Wykres funkcji x^2+5 dla {x_start}>x>{x_stop}")
    plt.xlabel("Wartość x")
    plt.ylabel("Wartość y")
    plt.legend("x^2+5")
square_function(-1,1)
square_function(-6,6)
square_function(0,5)

```
